```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Baseline Methods

```python
dummy_results = pd.read_csv("../data/Results-Data/dummy-performance.csv",index_col=0)
dummy_results.head()
```

```python
pd.DataFrame(dummy_results.mean())
```

```python
favorite_accuracy_tracker = np.array(dummy_results['Favorite Acc'])
underdog_accuracy_tracker = np.array(dummy_results['Underdog Acc'])
mixed_accuracy_tracker = np.array(dummy_results['Mixed Acc'])
favorite_ROI_tracker = np.array(dummy_results['Favorite ROI'])
underdog_ROI_tracker = np.array(dummy_results['Underdog ROI'])
mixed_ROI_tracker = np.array(dummy_results['Mixed ROI'])
```

```python
plt.figure(figsize=(12, 7))

plt.bar(['Bet Favorite', 'Bet Underdog', 'Bet Mixed'], 
        [np.mean(dummy_results['Favorite ROI']),np.mean(dummy_results['Underdog ROI']), np.mean(dummy_results['Mixed ROI'])],
        color=['C0', 'C1', 'C2'])

plt.title("ROI by Naive Betting Strategy", size = 16)
plt.ylabel("Return on Investment", size = 12)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax.boxplot([favorite_ROI_tracker, underdog_ROI_tracker, mixed_ROI_tracker])
ax.set_xticklabels(['Bet Favorite', 'Bet Underdog', 'Bet Mixed'], size = 12)
ax.set_ylabel("Return on Investment", size = 12)
ax.set_title("NCAA Tournament ROI by Naive Betting Strategy", size =16);
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax.boxplot([favorite_accuracy_tracker, underdog_accuracy_tracker, mixed_accuracy_tracker])
ax.set_xticklabels(['Bet Favorite', 'Bet Underdog', 'Bet Mixed'], size = 12)
ax.set_ylabel("Accuracy", size = 12)
ax.set_title("NCAA Tournament Accuracy by Naive Betting Strategy", size =16);
```

```python
list_of_years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018,
       2019, 2021]

plt.figure(figsize=(9, 5))

plt.plot(list_of_years, favorite_ROI_tracker, label = "Bet Favorite")
plt.plot(list_of_years, underdog_ROI_tracker, label = "Bet Underdog")
plt.plot(list_of_years, mixed_ROI_tracker, label = "Bet Mixed")
plt.title("NCAA Tournament ROI by Naive Betting Strategy", size = 16)
plt.xlabel("Year", size = 12)
plt.ylabel("Return on Investment", size = 12)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9, 5))

plt.plot(list_of_years, favorite_accuracy_tracker, label = "Bet Favorite")
plt.plot(list_of_years, underdog_accuracy_tracker, label = "Bet Underdog")
plt.plot(list_of_years, mixed_accuracy_tracker, label = "Bet Mixed")
plt.title("NCAA Tournament Accuracy by Naive Betting Strategy", size = 16)
plt.xlabel("Year", size = 12)
plt.ylabel("Accuracy", size = 12)
plt.legend()
plt.show()
```

# Machine Learning Models

```python
results_df = pd.read_csv("../data/Results-Data/model_performance.csv",index_col=0)
```

```python
# Average results
pd.DataFrame(results_df[['LR acc Accuracy Basic', 'LR acc ROI Basic',
           'KNN acc Accuracy Basic', 'KNN acc ROI Basic',
           'SVM acc Accuracy Basic', 'SVM acc ROI Basic',
           'LR acc Accuracy Enhanced', 'LR acc ROI Enhanced',
           'KNN acc Accuracy Enhanced', 'KNN acc ROI Enhanced',
           'SVM acc Accuracy Enhanced', 'SVM acc ROI Enhanced']].mean())
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax.boxplot([results_df['LR acc Accuracy Basic'], results_df['KNN acc Accuracy Basic'], results_df['SVM acc Accuracy Basic']])
ax.set_xticklabels(['LR', 'KNN', 'SVM'], size = 12)
ax.set_ylabel("Accuracy", size = 12)
ax.set_title("NCAA Tournament Accuracy by ML Model w/ Uniform Betting Strategy", size =16);
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax.boxplot([results_df['LR acc ROI Basic'], results_df['KNN acc ROI Basic'], results_df['SVM acc ROI Basic']])
ax.set_xticklabels(['LR', 'KNN', 'SVM'], size = 12)
ax.set_ylabel("Return on Investment", size = 12)
ax.set_title("NCAA Tournament ROI by ML Model w/ Uniform Betting Strategy", size =16);
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax.boxplot([results_df['LR acc Accuracy Enhanced'], results_df['KNN acc Accuracy Enhanced'], results_df['SVM acc Accuracy Enhanced']])
ax.set_xticklabels(['LR', 'KNN', 'SVM'], size = 12)
ax.set_ylabel("Accuracy", size = 12)
ax.set_title("NCAA Tournament Accuracy by ML Model w/ Edge Betting Strategy", size =16);
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax.boxplot([results_df['LR acc ROI Enhanced'], results_df['KNN acc ROI Enhanced'], results_df['SVM acc ROI Enhanced']])
ax.set_xticklabels(['LR', 'KNN', 'SVM'], size = 12)
ax.set_ylabel("Return on Investment", size = 12)
ax.set_title("NCAA Tournament ROI by ML Model w/ Edge Betting Strategy", size =16);
```

```python
# create data
x = np.arange(3)
basic_ROIs = [np.mean(results_df['LR acc ROI Basic']), 
              np.mean(results_df['SVM acc ROI Basic']), 
             np.mean(results_df['KNN acc ROI Basic'])]

enhanced_ROIs = [np.mean(results_df['LR acc ROI Enhanced']), 
              np.mean(results_df['SVM acc ROI Enhanced']), 
             np.mean(results_df['KNN acc ROI Enhanced'])]
width = 0.4

# plot data in grouped manner of bar type
plt.bar(x-0.4, basic_ROIs, width, color='#003ba3')
plt.bar(x, enhanced_ROIs, width, color='#4a8cff')
plt.xticks(x-0.5*width, ['LR', 'SVM', 'KNN'])
plt.ylabel("Avg. Return on Investment")
plt.title("ROI by Betting Strategy")
plt.legend(["Bet Uniform", "Bet Edge"])
plt.grid(color = 'lightgrey', linestyle = '--', linewidth = 0.5)
plt.axhline(y=0, linewidth=2, color="black")

plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(9, 5)

ax.bar(['Bet Favorite', 'Bet Underdog', 'Bet Mixed','LR Bet Edge','SVM Bet Edge','KNN Bet Edge'], 
        [np.mean(dummy_results['Favorite ROI']), 
         np.mean(dummy_results['Underdog ROI']),
         np.mean(dummy_results['Mixed ROI']),
         np.mean(results_df['LR acc ROI Enhanced']), 
         np.mean(results_df['SVM acc ROI Enhanced']), 
         np.mean(results_df['KNN acc ROI Enhanced'])
        ],
        color=['C0', 'C1', 'C2','purple','red','cyan'])

ax.grid(color = 'lightgrey', linestyle = '--', linewidth = 0.5)
ax.axhline(y=0, linewidth=2, color="black")
ax.set_axisbelow(True)


ax.set_title("ROI by Betting Strategy", size = 16)
ax.set_ylabel("Avg. Return on Investment", size = 12)
plt.show()
```
